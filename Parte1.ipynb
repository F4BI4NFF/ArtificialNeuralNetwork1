```python
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
X_train, y_train = make_moons(200, noise=0.2)

X_test, y_test = make_moons(500, noise=0.2)

plt.scatter(X_train[:,0], X_train[:,1], s=40, c=y_train, cmap=plt.cm.winter)
#plt.scatter(X_test[:,0], X_test[:,1], s=20, c=y_test, cmap=plt.cm.Spectral)
plt.show()
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD

n_h=1 #1 neurona

model = Sequential()
model.add(Dense(n_h, input_dim=X_train.shape[1], init="uniform", activation="sigmoid"))
model.compile(optimizer=SGD(lr=0.01), loss="binary_crossentropy")

model.fit(X_train, y_train, nb_epoch=20, batch_size=100, verbose=0)
print "\nTest Accuracy (activation sigmoid) = %f"%model.evaluate(X_test, y_test)
print "---------------------------------------------------------"

model = Sequential()
model.add(Dense(n_h, input_dim=X_train.shape[1], init="uniform", activation="softplus"))
model.compile(optimizer=SGD(lr=0.01), loss="binary_crossentropy")

model.fit(X_train, y_train, nb_epoch=20, batch_size=100, verbose=0)
print "\nTest Accuracy (activation softplus) = %f"%model.evaluate(X_test, y_test)
print "---------------------------------------------------------"

model = Sequential()
model.add(Dense(n_h, input_dim=X_train.shape[1], init="uniform", activation="tanh"))
model.compile(optimizer=SGD(lr=0.01), loss="binary_crossentropy")

model.fit(X_train, y_train, nb_epoch=20, batch_size=100, verbose=0)
print "\nTest Accuracy (activation tanh) = %f"%model.evaluate(X_test, y_test)
print "---------------------------------------------------------"

#mas arquitecturastpe
```
